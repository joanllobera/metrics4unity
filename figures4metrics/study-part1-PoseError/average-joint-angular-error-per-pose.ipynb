```python
import pandas as pd
import numpy as np
import pylab as plt


commonaxis = [20, 120, 10, 70]
```

```python


#df=pd.read_csv('distance-test-results.csv', sep=',',header=0)
df128M=pd.read_csv('run3011-Walk1041MujocoSPDDReConNoState/AssetsposeDifferenceDReCon-Mj-128000092.csv', sep=',')
df64M=pd.read_csv('run3011-Walk1041MujocoSPDDReConNoState/AssetsposeDifferenceDReCon-Mj-63999935.csv', sep=',')
df32M=pd.read_csv('run3011-Walk1041MujocoSPDDReConNoState/AssetsposeDifferenceDReCon-Mj-31999940.csv', sep=',')

df128M.columns
```

```python
plt.figure(figsize=(15, 10), dpi=300)

plt.plot(df128M['Time'], df128M['Sample'], '-r', label='128M')
plt.plot(df128M['Time'], np.mean(df128M['Sample'])*np.ones(df128M['Sample'].shape), '--r')

plt.plot(df64M['Time'], df64M['Sample'], '-g', label='64M')
plt.plot(df64M['Time'], np.mean(df64M['Sample'])*np.ones(df64M['Sample'].shape), '--g')

plt.plot(df32M['Time'], df32M['Sample'], '-b', label='32M')
plt.plot(df32M['Time'], np.mean(df32M['Sample'])*np.ones(df32M['Sample'].shape), '--b')

plt.legend(loc='upper right')
plt.axis(commonaxis)
#plt.errorbar(df.index,df['angle_mean'], yerr=df['angle_std'], c= "red", label="angle difference",marker="o", linewidth=0,elinewidth=1)

plt.title("Walk Mujoco SPD DReCon NoState \n Average joint angular error per pose ")
plt.xlabel("frame")
plt.ylabel("degrees ")
```
